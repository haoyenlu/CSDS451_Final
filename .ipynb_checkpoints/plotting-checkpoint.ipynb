```python
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

import numpy as np
import pandas as pd

import os
from glob import glob
```

```python
csvs = {}
csv_root = 'csv'
for file in sorted(glob(os.path.join(csv_root,'*.csv'))):
    csvs[os.path.splitext(os.path.basename(file))[0]] = pd.read_csv(file)
```

```python
layers = [x for x in csvs.keys()]
```

```python
methods = csvs[layers[0]].columns.tolist()
```

```python
avg = {x:{} for x in layers}
for layer in layers:
    for method in methods:
        avg[layer][method] = csvs[layer][method].mean()
```

```python
for i, method in enumerate(methods):
    x = layers
    y = []
    for layer in layers:
        y.append(avg[layer][method])
    plt.plot(x,y)
```
